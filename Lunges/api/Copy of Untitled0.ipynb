```python
def txttocsv2(path):
    num = 0
    folder_list = os.listdir(path)
    for folders in folder_list:
      print('Folder:', folders)
      if folders == 'Correct -0': continue
      file_list = os.listdir(path+folders)
      for file_ in file_list:
        num += 1
        if 'txt' not in file_: continue 
        if os.stat(path+folders+'/'+file_).st_size == 0: continue
        with open(path+folders+'/'+file_) as f:
            # if f.readlines() == []: return None
            # print(f.readlines())
            lines = f.readlines()[0]
        f.close()
        lt = lines.split('#')
        df = pd.DataFrame()
        for i in lt[:-1]:
            df = df.append([i.split(',')], ignore_index=True)
        df.set_axis(select, axis=1, inplace=True)
        if os.path.isdir('/content/drive/MyDrive/Fitbuddy_Susanket/csv_with_lib/lunges/csv_frames/'+folders) is False: os.mkdir('/content/drive/MyDrive/Fitbuddy_Susanket/csv_with_lib/lunges/csv_frames/'+folders)
        df.to_csv('/content/drive/MyDrive/Fitbuddy_Susanket/csv_with_lib/lunges/csv_frames/'+folders+'/frame_'+str(num)+'.csv',index=False)
        print('finished procesing: ', folders, file_)
```
